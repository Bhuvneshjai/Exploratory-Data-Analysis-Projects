```python
df_country = pd.read_excel('Country-Code.xlsx')
df_country.head()
```

```python
df.columns
```

```python
final_df = pd.merge(df, df_country, on='Country Code', how='left')
```

```python
final_df
```

```python
final_df.dtypes
```

```python
country_names=final_df.Country.value_counts().index
```

```python
country_val=final_df.Country.value_counts().values
```

```python
# Pie Chart - Top 3 countries that uses zomato
plt.pie(country_val[:3], labels=country_names[:3], autopct='%1.2f%%')
```

```python
# Observation: Zomato maximum records or transactions are from India after that USA and then United Kingdom
```

```python
final_df.columns
```

```python
final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text'])
```

```python
final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size()
```

```python
final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index()
```

```python
ratings = final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
```

```python
ratings
```

## Observations
1. When ratings are from 4.5 to 4.9 is Excellent
2. When ratings are from 4.0 to 4.4 is Very Good
3. When ratings are from 3.5 to 3.9 is Good
4. When ratings are from 2.5 to 3.4 is Average
5. When ratings are from 1.8 to 2.4 is Poor
6. When ratings are 0.0 is Not Rated

```python
plt.rcParams['figure.figsize'] = (12, 6)
sns.barplot(x='Aggregate rating', y='Rating Count', data=ratings)
```

```python
sns.barplot(x='Aggregate rating', y='Rating Count', hue='Rating color', data=ratings)
```

```python
sns.barplot(x='Aggregate rating', y='Rating Count', hue='Rating color', data=ratings, palette=['blue','red','orange','yellow','green','green'])
```

## Observation:
1. Not Rated count is very high
2. Maximum number of rating are between 2.5 to 3.4

```python
# Count Plot
sns.countplot(x="Rating color", data=ratings, palette=['blue','red','orange','yellow','green','green'])
```
